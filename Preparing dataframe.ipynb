```python
data_wiki["Deaths"] = pd.to_numeric(data_wiki["Deaths"].str.extract(r'^(\d+)', expand=False))
```

```python
data_wiki["Recovered"] = pd.to_numeric(data_wiki["Recovered"].str.extract(r'^(\d+)', expand=False))
```

```python
data_wiki["Cases"] = pd.to_numeric(data_wiki["Cases"].str.extract(r'^(\d+)', expand=False))
```

```python
def prepare_dataframe(data, n):
    new = pd.DataFrame()
    new["Cases"] = data.iloc[(n - 1):, 0].values
    new["Recovered"] = data.iloc[(n - 1):, 1].values
    new["Deaths"] = data.iloc[(n - 1):, 2].values
    for j in range(1, n):
        
        new["Cases " + str(j)+" days before"] = data.iloc[(n - j - 1):-j, 0].values
        new["Recovered " + str(j)+" days before"] = data.iloc[(n - j - 1):-j, 1].values
        new["Deaths " + str(j)+" days before"] = data.iloc[(n - j - 1):-j, 2].values
    return new
```

## The same but with target as a list

```python
def prepare_dataframe2(data, n):
    new = pd.DataFrame()
    targetColumn = []
    for row in range(n-1,len(data.index)):
        targetColumn.append([data.iloc[row, 0], data.iloc[row, 1],data.iloc[row, 2]])
    new['target'] = targetColumn
    
    for j in range(1, n):
        new["Cases " + str(j)+" days before"] = data.iloc[(n - j - 1):-j, 0].values
        new["Recovered " + str(j)+" days before"] = data.iloc[(n - j - 1):-j, 1].values
        new["Deaths " + str(j)+" days before"] = data.iloc[(n - j - 1):-j, 2].values
    
    return new
```
